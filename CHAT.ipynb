```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import random
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as img 
```

```python
import json
with open('context.json') as json_data:
    q = json.load(json_data)
```

```python
q
```

```python
a= q['intents']
a
```

```python
classes = []
X=[]
y=[]
responses = {}
con = []
context_to_set = {}
context_to_filter = {}
    
responses
for intent in a:
    con.append(intent['context_filter'])
    classes.append(intent['tag'])
    X.extend(intent['patterns'])
    y.extend([intent['tag']]*len(intent['patterns']))
    responses[intent['tag']] = intent['responses']
    context_to_set[intent['tag']]= intent['context_set']
    context_to_filter[intent['tag']]= intent['context_filter']
```

```python
context_to_filter
```

```python
context_to_set
```

```python
pipeline = make_pipeline(CountVectorizer(),LogisticRegression())
pipeline.fit(X,y)
```

```python
pipeline.classes_
```

```python
context = None
test = "Welcome to ZenRays"
new_test = test.center(70)
print(new_test)
while(True):
    print("\n")
    question = input("User:    ")
    classes_proba = pipeline.predict_proba([question])[0]
    answer_classes = [] 
    for index,cat in enumerate(pipeline.classes_):
        filters = context_to_filter[cat]
        
        if filters==None:
            answer_classes.append([cat,classes_proba[index]])
        elif context in filters:
            answer_classes.append([cat,classes_proba[index]])
        else:
            continue 
    answer_classes.sort(key=lambda x: x[1],reverse=True)  
    
    cat = answer_classes[0][0]
    print("ZenRays:",random.choice(responses[cat]))
    context = context_to_set[cat]
    
   
        
```
